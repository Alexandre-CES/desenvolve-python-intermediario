#### Q3. 
Crie um dataframe `participacao` com as colunas `país`, `copa`, e `num_copas` que armazena a quantidade de copas do mundo que cada país participou, informando se é da copa masculina ou feminina. Ao final imprima, como apresentado a seguir, o top 5 países de cada competição que mais participou de copas do mundo.

```

```
país            copa        num_copas

#### Q5. 
Qual país tomou mais cartões amarelos somando todas as copas?

Neste exercício você vai trabalhar com as colunas `cartao_amarelo_1` e `cartao_amarelo_2` onde cada observação é uma string represetando uma lista dos cartões amarelos de um único jogo na forma `minuto|placar|jogador(a)`. Por exemplo:
```
['16’|1:0|Rosana Gómez', '20’|2:0|Gabriela Chávez]
```

Recomendo criar colunas `num_cartoes_1` e `num_cartoes_2` a partir de cada coluna `cartao_amarelo_X` usando o método genérico `apply` para chamar uma função que você vai criar para transformar uma observação de cartão amarelo em uma contagem de elementos da lista. 

Lembre de levar em consideração que muitas observações são `NaN`. 

Em seguida faça sua mágica para agrupar as informações por país, acumular os cartões de jogos em casa e jogos visitante, e produzir o resultado final como apresentado a seguir (valores inventados):

```
país        cartões amarelos
Argentina   97
England     93
Australia   93
...
```

````python
def contar_cartoes(cartoes_str):

    #retornar se for NaN
    if pd.isna(cartoes_str):
        return 0
    
    #remover caracteres indesejados
    cartoes_str = cartoes_str.replace('&rsquor;', "'").replace("[", "").replace("]", "").replace("'", "")

    cartoes_str = cartoes_str.split(", ") #dividir cartões em forma de lista
    return len(cartoes_str) #retornar contagem de cartões

#pegar dados necessários e os dividindo
cartoes_1_df = wc[['cartao_amarelo_1', 'time_1']].rename(columns={'time_1':'pais','cartao_amarelo_1': 'cartoes amarelos'})
cartoes_2_df = wc[['cartao_amarelo_2', 'time_2']].rename(columns={'time_2':'pais','cartao_amarelo_2': 'cartoes amarelos'})

#concatena os dois 
cartoes_df = pd.concat([cartoes_1_df, cartoes_2_df])

#aplicar função na coluna de cartões amarelos
cartoes_df['cartoes amarelos'] = cartoes_df['cartoes amarelos'].apply(contar_cartoes)

#fazer novo dataframe
cartoes_amarelos = cartoes_df.groupby(['pais'])['cartoes amarelos'].sum().reset_index(name='cartões amarelos')

#mostrar na tela
display(cartoes_amarelos)
````

#### Q6.
Qual o top10 jogadores com mais gols em copa? Considere gols em jogo e gols de pênalti.

Para conseguir essa informação, você precisará trabalhar com as colunas:
```
10  gols_1_detalhes         
11  gols_2_detalhes    
14  gols_1_penalti     
15  gols_2_penalti     
```

Essas também são colunas textuais, onde cada observações apresenta todos os gols de uma partida separados pelo caracter `|`. Cada gol está na forma `'Jogador(a) · minuto’'`. Por exemplo:
```
'Alex Morgan · 12’|Rose Lavelle · 20’'
```

Lembre de levar em consideração que muitas observações são `NaN`. 

Recomendo criar um dicionário à parte, onde cada chave será um jogador encontrado nessas colunas do dataframe `wc` e o valor correspondente será a contagem de ocorrências desses nomes.

Em seguida basta converter o seu dicionário em um dataframe e imprimí-lo na forma (valores não são inventados 😁):
```
jogador(a)      num_gols 
Marta           17
Miroslav Klose  16
... 
```

````python
jogadores_e_gols = {}

def Main():

    #renomear todos para apenas gols
    gols_1_detalhes_df = wc[['gols_1_detalhes']].rename(columns={'gols_1_detalhes':'gols'})
    gols_2_detalhes_df = wc[['gols_2_detalhes']].rename(columns={'gols_2_detalhes':'gols'})
    gols_1_penalti_df = wc[['gols_1_penalti']].rename(columns={'gols_1_penalti':'gols'})
    gols_2_penalti_df = wc[['gols_2_penalti']].rename(columns={'gols_2_penalti':'gols'})

    #concatenar tudo em um só dataframe
    gols_df = pd.concat([gols_1_detalhes_df, gols_2_detalhes_df, gols_1_penalti_df, gols_2_penalti_df])
    
    #aplicar função em todos os itens
    gols_df['gols'].apply(processar_gols)

    #criar dataframe com resultado
    gols_df = criar_dataframe_jogadores()

    #mostrar na tela
    display(gols_df)

def processar_gols(gols_str):
    '''Divide os gols de cada partida'''
    #pular NaN
    if pd.isna(gols_str):
        return 0
    
    #dividir jogadores
    gols_str = gols_str.split('|')

    for item in gols_str:

        #deixar só o nome do jogador
        item_tmp = item
        if '·' in item:
            item_tmp = item.split(' · ')
            item_tmp = item_tmp[0]
        
        adicionar_ao_dicionario(item_tmp)
    
def adicionar_ao_dicionario(item):
    '''adiciona jogadores novos e atualiza os já existentes no dicionário'''
    if item in jogadores_e_gols:
        jogadores_e_gols[item] += 1
    else:
        jogadores_e_gols[item] = 1
    
def criar_dataframe_jogadores():
    '''Cria um novo dataframe com um dicionário de jogadores/gols'''

    df_jogadores = pd.DataFrame(list(jogadores_e_gols.items()), columns=['jogador', 'numero_de_gols'])

    return df_jogadores

Main()
````
